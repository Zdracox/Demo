```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot([1, 2, 3, 4], [1, 4, 9, 16])
ax1.set_title("Subplot 1(line)")

ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter([1, 2, 3, 4], [1, 2, 3, 4], color = 'red')
ax2.set_title("Subplot 2(scatter)")

ax3 = fig.add_subplot(2, 2, 3)
ax3.bar(['A', 'B', 'C', 'D'], [10, 20, 15, 30], color = 'green')
ax3.set_title("Subplot 3(bar)")

ax4 = fig.add_subplot(2, 2, 4)
ax4.hist([3, 4, 5, 6, 6, 6, 7, 8, 9], bins = 5, color = 'purple')
ax4.set_title("Subplot 4(hist)")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, sharex = True, figsize = (6,4))
axs[0].plot([1, 2, 3, 4], [1, 4, 9, 16], marker = 'o', color = 'blue')
axs[0].set_title('Subplot 1(Line)')
axs[0].set_ylabel('Y-axis for Subplot 1')

axs[1].scatter([1, 2, 3, 4], [0, 3, 2, 4], marker = 'x', color = 'red')
axs[0].set_title('Subplot 2(Scatter)')
axs[0].set_ylabel('Y-axis for Subplot 1')

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]
fig, ax = plt.subplots(figsize = (8, 4))

ax.plot(x, y, label = 'Line Plot', marker = 'o', linestyle = '-')

ax.bar(x, y, width = 0.3, label = 'Bar plot', align = 'center', alpha = 0.6, color = 'green')

ax.scatter(x, y, label = 'Scatter plot', color = 'red', marker = 'x', s = 50)

ax.set_title('Different chart types on the same axis')
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

ax.legend()
plt.show()
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 20]

x2 = [1, 2, 3, 4, 5]
y2 = [5, 8, 16, 2, 15]

fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(x1, y1, label = 'Line Chart 1', marker  = 'o', linestyle = '-', color = 'brown')

ax.plot(x2, y2, label = 'Line Chart 2', marker = 's', linestyle = '--', color = 'yellow')

ax.set_title ('Two line charts in the same plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.legend(loc = 'best')
plt.show()
```
